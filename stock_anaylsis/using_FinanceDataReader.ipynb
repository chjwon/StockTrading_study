```python
# check alpha of price in btc between krw and usd
usd_krw = fdr.DataReader('USD/KRW','2016-01-01')
btc_krw = fdr.DataReader('BTC/KRW','2016-01-01') # bithumb
btc_usd = fdr.DataReader('BTC/USD','2016-01-01') # bitfinex


print("old usd_krw columns",usd_krw.columns)
print("old btc_krw columns",btc_krw.columns)
print("old btc_usd columns",btc_usd.columns)

usd_krw.columns = ['Close_usd_krw', 'Open_usd_krw', 'High_usd_krw', 'Low_usd_krw', 'Change_usd_krw']
btc_krw.columns = ['Close_btc_krw', 'Open_btc_krw', 'High_btc_krw', 'Low_btc_krw', 'Volume_btc_krw', 'Change_btc_krw']
btc_usd.columns = ['Close_btc_usd', 'Open_btc_usd', 'High_btc_usd', 'Low_btc_usd', 'Volume_btc_usd', 'Change_btc_usd']


print("new usd_krw columns",usd_krw.columns)
print("new btc_krw columns",btc_krw.columns)
print("new btc_usd columns",btc_usd.columns)

usd_krw
```

```python
temp = pd.merge(btc_krw,btc_usd, how='outer',on='Date')
temp.head()
```

```python
df = pd.merge(temp,usd_krw,how='outer',on='Date')
df.head()
```

```python
df = df.fillna(method='ffill')
```

```python
df.columns
```

```python
df = df[['Close_btc_krw','Close_btc_usd','Close_usd_krw']]
df
```

```python
df['Close_btc_krw*usd'] = df['Close_btc_usd'] * df['Close_usd_krw']
df
```

```python
df['Close_btc_krw*usd'].plot()
df['Close_btc_krw'].plot()
```
